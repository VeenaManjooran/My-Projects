```python
#creating a new df from existing df with selected columns
df2 = df[['neighborhood','size_in_sqft','no_of_bedrooms','price','quality','security','view_of_water','private_pool',]].copy()
df2.head()
```

```python
df2.to_csv("Property_data")
```

```python
dummies = pd.get_dummies(df2.neighborhood)
dummies
```

```python
merged = pd.concat([df2,dummies],axis='columns')
merged
```

```python
final = merged.drop(['neighborhood','Mina Rashid'], axis='columns')
final
```

```python
X=final.drop(columns=['price'])
y=final['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
print(X_train.shape)
print(y_train.shape)
```

# Applying Linear Regression

```python
# Define the column transformer without OneHotEncoder
column_trans = make_column_transformer(
    (StandardScaler(), ['size_in_sqft', 'no_of_bedrooms']),
    remainder='passthrough'
)
```

```python
scaler = StandardScaler()
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

# X being feature matrix
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
pipe1 = make_pipeline(column_trans,scaler, lr)
```

```python
pipe1.fit(X_train,y_train)
```

```python
y_pred_lr = pipe1.predict(X_test)
```

```python
r2_score(y_test,y_pred_lr)
```

# lasso


```python
lasso = Lasso() 
```

```python
pipe2 = make_pipeline(column_trans,scaler, lasso)
```

```python
pipe2.fit(X_train,y_train)
```

```python
y_pred_lasso = pipe2.predict(X_test)
r2_score(y_test,y_pred_lasso)
```

# using Ridge

```python
ridge = Ridge()
```

```python
pipe3 = make_pipeline(column_trans,scaler, ridge)
```

```python
pipe3.fit(X_train,y_train)
```

```python
y_pred_ridge = pipe3.predict(X_test)
r2_score(y_test,y_pred_ridge)
```
